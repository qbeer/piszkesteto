```python
from astropy.io import fits
import numpy as np
```

```python
red_channel = fits.open('./masterR.fits')
green_channel = fits.open('./masterG.fits')
blue_channel = fits.open('./masterB.fits')
```

```python
r = red_channel[0].data
g = green_channel[0].data
b = blue_channel[0].data

r.shape, g.shape, b.shape, np.max(r), np.max(g), np.max(b), np.mean(r), np.median(g), np.median(b)
```

```python
rgbArray = np.zeros((2048, 2048, 3), 'uint8')
rgbArray[..., 0] = 255.*r/1800.
rgbArray[..., 1] = 255.*g/2000.
rgbArray[..., 2] = 255.*b/1500.
```

```python
import matplotlib.pyplot as plt
```

```python
rgbArray.shape
```

```python
plt.figure(figsize=(9, 9))
plt.imshow(r/1800., vmin=0, vmax=1, cmap=plt.cm.Reds)
```

```python
plt.figure(figsize=(9, 9))
plt.imshow(g/2000., vmin=0, vmax=1, cmap=plt.cm.Greens)
```

```python
plt.figure(figsize=(9, 9))
plt.imshow(b/1500., vmin=0, vmax=1., cmap=plt.cm.Blues)
```

```python
plt.imshow(rgbArray, vmin=0, vmax=255, interpolation='none', )
plt.axis('off')
plt.show()
```

```python
from astropy.visualization import make_lupton_rgb
```

```python
rgb_img = make_lupton_rgb(230.*r/255., 249.*g/255., b, Q=25, stretch=1.1)
```

```python
plt.figure(figsize=(9, 9))
plt.imshow(rgb_img, origin='lower')
```
